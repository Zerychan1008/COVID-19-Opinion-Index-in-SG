### Sentiment Analysis

```python
import time
from textblob import TextBlob
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
data_path = "/content/drive/My Drive/covid19sg_project/datafolder/tweets/"
numday=182
date='2020-01-21'
s_date = datetime.strptime(date, "%Y-%m-%d")
senti_scores = []
dates = []
hot_topics = []
for idx in range(numday):
    e_date = s_date + timedelta(days=idx)
    end_date = datetime.strftime(e_date, "%Y-%m-%d")
    df_all = pd.read_csv(data_path + "{}.csv".format(end_date))
    corpus = df_all.Text.tolist()
    scores = []
    for sentext in corpus:
        senti = TextBlob(sentext)
        score = senti.sentiment.polarity
        scores.append(score)
    hot_topics.append(len(scores))
    senti_scores.append(np.mean(scores)) 
    dates.append(end_date)
```

```python
df_result = pd.DataFrame()
```

```python
df_result['Senti'] = senti_scores
df_result['datetime'] = dates
df_result['attention']  = hot_topics
```
